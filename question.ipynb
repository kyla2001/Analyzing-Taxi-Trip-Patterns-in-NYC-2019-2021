# Data Preprocessing

```python
import pandas as pd
import glob
import os
import natsort
from pathlib import Path
from matplotlib import pyplot as plt
```

```python
base_data = Path("../data")
years = ["2019", "2020", "2021"]
```

```python
# merge defined function
def merge_parquet_files(data_folder, file_prefix):
    full_data_folder = os.path.join(base_data, data_folder)
    
    file_pattern = os.path.join(full_data_folder, f"{file_prefix}*.parquet")
    file_list = glob.glob(file_pattern)
    
    file_list = natsort.natsorted(file_list)

    combined_df = pd.concat([pd.read_parquet(file) for file in file_list], ignore_index=True)
    return combined_df


yellow_2019 = yellow_2020 = yellow_2021 = None
green_2019 = green_2020 = green_2021 = None

for year in years:

    yellow_df = merge_parquet_files(year, "yellow_tripdata_")
    green_df = merge_parquet_files(year, "green_tripdata_")

    if yellow_df is not None:
        globals()[f"yellow_{year}"] = yellow_df
        print(f"{year} Yellow Taxi Data：")
        display(yellow_df.head())

    if green_df is not None:
        globals()[f"green_{year}"] = green_df
        print(f"{year} Green Taxi Data：")
        display(green_df.head())
```

```python
yellow_2019 = yellow_2019.dropna(subset=['fare_amount', 'trip_distance', 'total_amount', 'passenger_count'])
yellow_2020 = yellow_2020.dropna(subset=['fare_amount', 'trip_distance', 'total_amount', 'passenger_count'])
yellow_2021 = yellow_2021.dropna(subset=['fare_amount', 'trip_distance', 'total_amount', 'passenger_count'])
green_2019 = green_2019.dropna(subset=['fare_amount', 'trip_distance', 'total_amount', 'passenger_count'])
green_2020 = green_2020.dropna(subset=['fare_amount', 'trip_distance', 'total_amount', 'passenger_count'])
green_2021 = green_2021.dropna(subset=['fare_amount', 'trip_distance', 'total_amount', 'passenger_count'])
```

### **B1. Heat Map Intensity**

```python
# step 1 - merge all taxi data

# dict to store files
files_dict = {
    "yellow": [],
    "green": []
}

# loop through years
for year in years:
    year_path = os.path.join(base_data, year)
    files = glob(os.path.join(year_path, "*.parquet"))
    for file in files:
        if "yellow_tripdata" in file:
            files_dict["yellow"].append(file)
        elif "green_tripdata" in file:
            files_dict["green"].append(file)

# check number of files
print("Yellow Taxi Files:", len(files_dict["yellow"]))
print("Green Taxi Files:", len(files_dict["green"]))
```

```python
# step 2 - calculate pickup and dropoff counts

def process_tripdata(files, taxi_type):
    pickup_list = []
    dropoff_list = []

    for file_path in files:
        # keep only pickup and dropoff location columns
        df = pd.read_parquet(file_path, engine='pyarrow')[['PULocationID', 'DOLocationID']]

        # extract time from file name
        time_str = file_path.split('/')[-1].split('_')[2].split('.')[0] 
        df['time'] = time_str

        # calculate pickup and dropoff counts
        pickup_counts = df['PULocationID'].value_counts().reset_index()
        pickup_counts.columns = ['location_i', 'count']
        pickup_counts['location_i'] = pickup_counts['location_i'].astype(float).round(1)
        pickup_counts['count'] = pickup_counts['count'].round(1)
        pickup_counts['time'] = time_str
        pickup_list.append(pickup_counts)

        dropoff_counts = df['DOLocationID'].value_counts().reset_index()
        dropoff_counts.columns = ['location_i', 'count']
        dropoff_counts['location_i'] = dropoff_counts['location_i'].astype(float).round(1)
        dropoff_counts['count'] = dropoff_counts['count'].round(1)
        dropoff_counts['time'] = time_str
        dropoff_list.append(dropoff_counts)

    # merge all months data
    df_pickup_counts = pd.concat(pickup_list, ignore_index=True)
    df_dropoff_counts = pd.concat(dropoff_list, ignore_index=True)

    return df_pickup_counts, df_dropoff_counts

# yellow taxi, green taxi pickup and dropoff counts
yellow_pickup_counts, yellow_dropoff_counts = process_tripdata(files_dict['yellow'], 'yellow')
green_pickup_counts, green_dropoff_counts = process_tripdata(files_dict['green'], 'green')

# check
print("Yellow Pickup Counts Data:")
yellow_pickup_counts.head()
```

```python
# remove zero & NA values
def clean_dataframe(df):
    return df[(df['location_i'] != 0) & (df['time'] != 0)]

yellow_pickup_counts= clean_dataframe(yellow_pickup_counts)
yellow_dropoff_counts= clean_dataframe(yellow_dropoff_counts)
green_pickup_counts= clean_dataframe(green_pickup_counts)
green_dropoff_counts= clean_dataframe(green_dropoff_counts)

# check
print("Cleaned Yellow Pickup Counts Data:")
yellow_pickup_counts.head()
```

```python
# step 3 - add zone information from taxi_zones.csv

# read taxi_zones.csv
zones = pd.read_csv('../data/taxi_zones.csv')

# keep 'LocationID', 'zone', 'borough' columns
zones = zones[['LocationID', 'zone', 'borough']]

# rename 'LocationID' to 'location_i'
zones.columns = ['location_i', 'zone', 'borough']
zones['location_i'] = zones['location_i'].astype(float).round(1)


# add zone information to pickup and dropoff counts
def add_zone_info(df_heatmap):
    return df_heatmap.merge(zones, on='location_i', how='left', suffixes=('', '_drop'))

yellow_pickup_heatmap = add_zone_info(yellow_pickup_counts)
yellow_dropoff_heatmap = add_zone_info(yellow_dropoff_counts)
green_pickup_heatmap = add_zone_info(green_pickup_counts)
green_dropoff_heatmap = add_zone_info(green_dropoff_counts)

# check
print("Yellow Pickup Heatmap Data:")
yellow_pickup_heatmap.head()
```

```python
# step 4 - match gdf['location_i'] with heatmap_data['location_i'] and create heatmap

# create monthly heatmap based on geojson data
def create_monthly_heatmap(gdf, heatmap_data, taxi_type, heatmap_type):
    # calculate max count for color scale
    max_count = heatmap_data['count'].max()

    unique_months = heatmap_data['time'].unique()
    for month in unique_months:
        # filter data for each month
        month_data = heatmap_data[heatmap_data['time'] == month]

        # merge gdf with heatmap_data based on 'location_i'
        gdf_month = gdf.merge(month_data, on='location_i', how='left')
        gdf_month['count'] = gdf_month['count'].fillna(0)

        # create a folium map, centered at the mean latitude and longitude of the data
        m = folium.Map(location=[gdf_month.geometry.centroid.y.mean(), gdf_month.geometry.centroid.x.mean()], zoom_start=11, tiles='CartoDB positron', attr='CartoDB positron')

        # use Choropleth to create heatmap
        choropleth = folium.Choropleth(
            geo_data=gdf_month,
            data=gdf_month,
            columns=['location_i', 'count'],
            key_on='feature.properties.location_i',
            fill_color='PuBuGn' if taxi_type == 'green' else 'YlOrBr',  
            fill_opacity=0.7,
            line_opacity=0.2,
            threshold_scale=[0, max_count/15, 2*max_count/15, 3*max_count/15, 4*max_count/15, 5*max_count/15, 6*max_count/15, 7*max_count/15, 8*max_count/15, 9*max_count/15, 10*max_count/15, 11*max_count/15, 12*max_count/15, 13*max_count/15, 14*max_count/15, max_count],
            legend_name=f'{taxi_type.capitalize()} Taxi {heatmap_type.capitalize()} Heat Map Intensity ({month})'
        ).add_to(m)

        # add legend to the map
        legend_html = '''
        <div style="position: fixed; 
                    bottom: 50px; right: 50px; width: 300px; height: 90px; 
                    background-color: white; z-index:9999; font-size:20px;
                    border:2px solid grey; padding: 10px;">
            <b>{} Taxi {} Heat Map Intensity ({})</b>
            <br>
        </div>
        '''.format(taxi_type.capitalize(), heatmap_type.capitalize(), month, max_count)

        m.get_root().html.add_child(folium.Element(legend_html))

        # save the map as html
        m.save(f"../results_geo/heatmap_html/{taxi_type}_{heatmap_type}_{month}_heatmap.html")

# read NYC Taxi Zones Shapefile
gdf = gpd.read_file("../data/NYC Taxi Zones")

# create monthly heatmap
create_monthly_heatmap(gdf, yellow_pickup_heatmap, 'yellow', 'pickup')
create_monthly_heatmap(gdf, yellow_dropoff_heatmap, 'yellow', 'dropoff')
create_monthly_heatmap(gdf, green_pickup_heatmap, 'green', 'pickup')
create_monthly_heatmap(gdf, green_dropoff_heatmap, 'green', 'dropoff')
```

```python
# step 5 - convert png files to gif

# get heatmap images from directories
def get_images_from_directories(directories):
    images = []
    for directory in directories:
        images += [Image.open(os.path.join(directory, f)) for f in sorted(os.listdir(directory)) if f.endswith(".png")]
    return images

# save images to gif
def save_images_to_gif(images, output_path):
    if images:
        images[0].save(output_path, save_all=True, append_images=images[1:], duration=350, loop=0)

# dictionary
directory_groups = {
    "green_pickup_heatmap": [
        "../results_geo/heatmap_png/green_pickup_2019",
        "../results_geo/heatmap_png/green_pickup_2020",
        "../results_geo/heatmap_png/green_pickup_2021"
    ],
    "green_dropoff_heatmap": [
        "../results_geo/heatmap_png/green_dropoff_2019",
        "../results_geo/heatmap_png/green_dropoff_2020",
        "../results_geo/heatmap_png/green_dropoff_2021"
    ],
    "yellow_pickup_heatmap": [
        "../results_geo/heatmap_png/yellow_pickup_2019",
        "../results_geo/heatmap_png/yellow_pickup_2020",
        "../results_geo/heatmap_png/yellow_pickup_2021"
    ],
    "yellow_dropoff_heatmap": [
        "../results_geo/heatmap_png/yellow_dropoff_2019",
        "../results_geo/heatmap_png/yellow_dropoff_2020",
        "../results_geo/heatmap_png/yellow_dropoff_2021"
    ]
}

# create gif for each group
for gif_name, directories in directory_groups.items():
    images = get_images_from_directories(directories)
    output_path = f"../results_geo/{gif_name}.gif"
    save_images_to_gif(images, output_path)
```

### **B2. Top 5 {taxi_type}_{heatmap_type} Zones**

```python
# Step 1 - Calculate top 5 zones for each year

def get_top5_zones_per_year(df, year):
    # Ensure 'time' column is of datetime type
    if not pd.api.types.is_datetime64_any_dtype(df['time']):
        df['time'] = pd.to_datetime(df['time'])
        
    # filter data from years
    df_filtered = df[df['time'].dt.year == year]
    # group by 'zone' and 'borough' and sum 'count'
    grouped = df_filtered.groupby(['zone', 'borough'], as_index=False)['count'].sum()
    # add 'year' column
    grouped['year'] = year
    # sort by 'count' in descending order and get top 5 zones
    top5_zones = grouped.sort_values(by='count', ascending=False).head(5)
    return top5_zones

years = [2019, 2020, 2021]

# define datasets to be used
datasets = {
    'Yellow Taxi Pickup Zones': yellow_pickup_heatmap,
    'Yellow Taxi Dropoff Zones': yellow_dropoff_heatmap,
    'Green Taxi Pickup Zones': green_pickup_heatmap,
    'Green Taxi Dropoff Zones': green_dropoff_heatmap
}

# get top 5 pickup / dropoff zones for each year
top5_results = {}
for dataset_name, df in datasets.items():
    top5_results[dataset_name] = {}
    for year in years:
        top5_results[dataset_name][year] = get_top5_zones_per_year(df, year)

# print
for dataset_name, yearly_data in top5_results.items():
    for year, top5_df in yearly_data.items():
        print(f"Top 5 zones for {dataset_name} in {year}:")
        print(top5_df[['zone', 'borough', 'count', 'year']])
        print("\n")

# merge all top 5 zones for 3 years
df_yellow_pickup = pd.concat([top5_results['Yellow Taxi Pickup Zones'][year] for year in years], ignore_index=True)

df_yellow_dropoff = pd.concat([top5_results['Yellow Taxi Dropoff Zones'][year] for year in years], ignore_index=True)

df_green_pickup = pd.concat([top5_results['Green Taxi Pickup Zones'][year] for year in years], ignore_index=True)

df_green_dropoff = pd.concat([top5_results['Green Taxi Dropoff Zones'][year] for year in years], ignore_index=True)
```

```python
# step 2 - plot top 5 zones

# Yellow Pickup Plot
pivot_yellow_pickup = df_yellow_pickup.pivot(index='zone', columns='year', values='count').fillna(0)
plt.figure(figsize=(8, 12))
pivot_yellow_pickup.plot(kind='barh', stacked=True, cmap='viridis', edgecolor='black')
plt.title('Top 5 Yellow Taxi Pickup Zones')
plt.xlabel('Count')
plt.ylabel(' ')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()

# Yellow Dropoff Plot
pivot_yellow_dropoff = df_yellow_dropoff.pivot(index='zone', columns='year', values='count').fillna(0)
plt.figure(figsize=(8, 12))
pivot_yellow_dropoff.plot(kind='barh', stacked=True, cmap='summer', edgecolor='black')
plt.title('Top 5 Yellow Taxi Dropoff Zones')
plt.xlabel('Count')
plt.ylabel(' ')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()

# Green Pickup Plot
pivot_green_pickup = df_green_pickup.pivot(index='zone', columns='year', values='count').fillna(0)
plt.figure(figsize=(8, 12))
pivot_green_pickup.plot(kind='barh', stacked=True, cmap='viridis', edgecolor='black')
plt.title('Top 5 Green Taxi Pickup Zones')
plt.xlabel('Count')
plt.ylabel(' ')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()

# Green Dropoff Plot
pivot_green_dropoff = df_green_dropoff.pivot(index='zone', columns='year', values='count').fillna(0)
plt.figure(figsize=(8, 12))
pivot_green_dropoff.plot(kind='barh', stacked=True, cmap='summer', edgecolor='black')
plt.title('Top 5 Green Taxi Dropoff Zones')
plt.xlabel('Count')
plt.ylabel(' ')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
import pmdarima as pm

def prepare_time_series(data, value_column):
    data['year_month'] = data['tpep_pickup_datetime'].dt.to_period('M')
    ts = data.groupby('year_month')[value_column].sum().reset_index()
    ts['year_month'] = ts['year_month'].dt.to_timestamp()
    return ts.set_index('year_month')

# Time series data
trip_distance_2019 = prepare_time_series(yellow_2019, 'trip_distance')
trip_distance_2020 = prepare_time_series(yellow_2020, 'trip_distance')
trip_distance_2021 = prepare_time_series(yellow_2021, 'trip_distance')

# SARIMA
sarima_model = pm.auto_arima(trip_distance_2019, seasonal=True, m=12, trace=True, 
                             error_action='ignore', suppress_warnings=True)
print(sarima_model.summary())

# Predict 2020 trip distance
forecast_2020 = sarima_model.predict(n_periods=len(trip_distance_2020))
forecast_index = trip_distance_2020.index

# Plot Forecast & Observed
plt.figure(figsize=(12, 6))
plt.plot(trip_distance_2020.index, trip_distance_2020['trip_distance'], label='Observed - 2020', marker='o')
plt.plot(forecast_index, forecast_2020, label='Forecast - 2020', marker='o', linestyle='--')
plt.title('Trip Distance Forecast vs Observed - 2020')
plt.ylabel('Trip Distance')
plt.xlabel('Year-Month')
plt.xlim(pd.Timestamp('2020-01-01'), pd.Timestamp('2020-12-31'))  # Set x-axis limits using datetime
plt.legend()
plt.show()
```

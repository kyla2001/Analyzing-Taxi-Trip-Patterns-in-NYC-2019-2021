# Data Preprocessing

```python
import pandas as pd
import glob
import os
import natsort
from pathlib import Path
from matplotlib import pyplot as plt
```

```python
base_data = Path("../data")
years = ["2019", "2020", "2021"]
```

```python
# merge defined function
def merge_parquet_files(data_folder, file_prefix):
    full_data_folder = os.path.join(base_data, data_folder)
    
    file_pattern = os.path.join(full_data_folder, f"{file_prefix}*.parquet")
    file_list = glob.glob(file_pattern)
    
    file_list = natsort.natsorted(file_list)

    combined_df = pd.concat([pd.read_parquet(file) for file in file_list], ignore_index=True)
    return combined_df


yellow_2019 = yellow_2020 = yellow_2021 = None
green_2019 = green_2020 = green_2021 = None

for year in years:

    yellow_df = merge_parquet_files(year, "yellow_tripdata_")
    green_df = merge_parquet_files(year, "green_tripdata_")

    if yellow_df is not None:
        globals()[f"yellow_{year}"] = yellow_df
        print(f"{year} Yellow Taxi Data：")
        display(yellow_df.head())

    if green_df is not None:
        globals()[f"green_{year}"] = green_df
        print(f"{year} Green Taxi Data：")
        display(green_df.head())
```

```python
yellow_2019 = yellow_2019.dropna(subset=['fare_amount', 'trip_distance', 'total_amount', 'passenger_count'])
yellow_2020 = yellow_2020.dropna(subset=['fare_amount', 'trip_distance', 'total_amount', 'passenger_count'])
yellow_2021 = yellow_2021.dropna(subset=['fare_amount', 'trip_distance', 'total_amount', 'passenger_count'])
green_2019 = green_2019.dropna(subset=['fare_amount', 'trip_distance', 'total_amount', 'passenger_count'])
green_2020 = green_2020.dropna(subset=['fare_amount', 'trip_distance', 'total_amount', 'passenger_count'])
green_2021 = green_2021.dropna(subset=['fare_amount', 'trip_distance', 'total_amount', 'passenger_count'])
```

```python
for df in [yellow_2019, yellow_2020, yellow_2021]:
    df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
    df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
# Function to calculate monthly rides
def monthly_ridership(df):
    df['month'] = df['tpep_pickup_datetime'].dt.to_period('M')
    return df.groupby('month').size()

rides_2019 = monthly_ridership(yellow_2019)
rides_2020 = monthly_ridership(yellow_2020)
rides_2021 = monthly_ridership(yellow_2021)

# Plotting the monthly ridership trends
plt.figure(figsize=(12, 6))
rides_2019.plot(label='2019')
rides_2020.plot(label='2020')
rides_2021.plot(label='2021')
plt.xlabel('Month')
plt.ylabel('Number of Rides')
plt.title('Monthly Ridership Trends (2019-2021)')
plt.legend()
plt.show()
```

```python
# Calculate average passengers per trip for each year
avg_passengers_2019 = yellow_2019['passenger_count'].mean()
avg_passengers_2020 = yellow_2020['passenger_count'].mean()
avg_passengers_2021 = yellow_2021['passenger_count'].mean()

print("Average Passengers per Trip:")
print("2019:", avg_passengers_2019)
print("2020:", avg_passengers_2020)
print("2021:", avg_passengers_2021)
```

```python
# Calculate average fare amount and total amount per trip for each year
for year, data in zip(['2019', '2020', '2021'], [yellow_2019, yellow_2020, yellow_2021]):
    avg_fare = data['fare_amount'].mean()
    avg_total = data['total_amount'].mean()
    avg_tip = data['tip_amount'].mean()
    print(f"Year {year} - Average Fare: {avg_fare}, Average Total: {avg_total}, Average Tip: {avg_tip}")
```

```python
# Function to calculate monthly revenue
def monthly_revenue(df):
    df['month'] = df['tpep_pickup_datetime'].dt.to_period('M')
    return df.groupby('month')['total_amount'].sum()

revenue_2019 = monthly_revenue(yellow_2019)
revenue_2020 = monthly_revenue(yellow_2020)
revenue_2021 = monthly_revenue(yellow_2021)

# Plotting monthly revenue trends
plt.figure(figsize=(12, 6))
revenue_2019.plot(label='2019')
revenue_2020.plot(label='2020')
revenue_2021.plot(label='2021')
plt.xlabel('Month')
plt.ylabel('Total Revenue ($)')
plt.title('Monthly Revenue Trends (2019-2021)')
plt.legend()
plt.show()
```

```python
# Calculate average surcharges per trip for each year
for year, data in zip(['2019', '2020', '2021'], [yellow_2019, yellow_2020, yellow_2021]):
    avg_congestion = data['congestion_surcharge'].mean()
    avg_improvement = data['improvement_surcharge'].mean()
    avg_mta_tax = data['mta_tax'].mean()
    print(f"Year {year} - Avg Congestion Surcharge: {avg_congestion}, Avg Improvement Surcharge: {avg_improvement}, Avg MTA Tax: {avg_mta_tax}")
```

```python
# Calculate average trip distance for each year
for year, data in zip(['2019', '2020', '2021'], [yellow_2019, yellow_2020, yellow_2021]):
    avg_distance = data['trip_distance'].mean()
    print(f"Year {year} - Average Trip Distance: {avg_distance} miles")
```

```python
# Top 10 pickup and drop-off locations for each year
for year, data in zip(['2019', '2020', '2021'], [yellow_2019, yellow_2020, yellow_2021]):
    top_pu_locations = data['PULocationID'].value_counts().head(10)
    top_do_locations = data['DOLocationID'].value_counts().head(10)
    print(f"\nYear {year} - Top Pickup Locations:\n", top_pu_locations)
    print(f"Year {year} - Top Drop-off Locations:\n", top_do_locations)
```

```python
# Compare average fare and trip distance between Yellow and Green taxis for 2019, 2020, and 2021
for year, yellow_data, green_data in zip(['2019', '2020', '2021'], [yellow_2019, yellow_2020, yellow_2021], [green_2019, green_2020, green_2021]):
    avg_yellow_fare = yellow_data['fare_amount'].mean()
    avg_green_fare = green_data['fare_amount'].mean()
    avg_yellow_distance = yellow_data['trip_distance'].mean()
    avg_green_distance = green_data['trip_distance'].mean()
    print(f"Year {year} - Yellow Taxi: Avg Fare: {avg_yellow_fare}, Avg Distance: {avg_yellow_distance}")
    print(f"Year {year} - Green Taxi: Avg Fare: {avg_green_fare}, Avg Distance: {avg_green_distance}")
```

```python
# Payment type distribution for each year
for year, data in zip(['2019', '2020', '2021'], [yellow_2019, yellow_2020, yellow_2021]):
    payment_dist = data['payment_type'].value_counts(normalize=True) * 100
    print(f"\nYear {year} - Payment Type Distribution (%)\n", payment_dist)
```
